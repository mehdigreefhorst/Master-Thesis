### Retrieving the possible architecture fields

```python
from collections import defaultdict
from typing import List, Dict


def count_architecture_types(model_list: List[Dict]):
    possible_architecture = defaultdict(set)

    for model in model_list:
        model_architecture = model["architecture"]
        possible_architecture["output_modalities"].update(model_architecture["output_modalities"])
        possible_architecture["input_modalities"].update(model_architecture["input_modalities"])
        possible_architecture["tokenizers"].add(model_architecture["tokenizer"])
        possible_architecture["modality"].add(model_architecture["modality"])
        possible_architecture["instruct_types"].add(model_architecture["instruct_type"])
    
        
    return possible_architecture
count_architecture_types(response["models"])
```

```python
### Retrieving the possible supported parameters of all models
```

```python
def get_all_supported_parameters(model_list: List[Dict]):
    supported_parameters = set()
    for model in model_list:
        supported_parameters.update(model["supported_parameters"])
    return supported_parameters

get_all_supported_parameters(response["models"])
```

```python
def get_pricing_keys(model_list: List[Dict]):
  pricing_keys = set()
  for model in model_list:
      pricing_keys.update(model["pricing"].keys())
  return pricing_keys

get_pricing_keys(response["models"])
```
